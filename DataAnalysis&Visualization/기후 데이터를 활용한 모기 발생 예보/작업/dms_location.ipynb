```python
import folium

# 지도의 중심을 지정하기 위해 위도와 경도의 평균 
lat = dms['위도'].mean()
long = dms['경도'].mean()

# 지도 띄우기
m = folium.Map([lat, long], zoom_start=12)

m
```

```python
for i in dms.index:
    sub_lat = dms.loc[i, '위도']
    sub_long = dms.loc[i, '경도']
    
    title = dms.loc[i, 'DMS 지역명']
    
    # 지도에 데이터 찍어서 보여주기
    folium.Marker([sub_lat, sub_long], tooltip=title).add_to(m)

# 한글이 나오지 않는 오류로 html로 trouble shooting
m.save('dms.html')

m
```

```python
# 지도 새로 띄우기
m = folium.Map([lat, long], zoom_start=12)

for i in dms.index:
    sub_lat = dms.loc[i, '위도']
    sub_long = dms.loc[i, '경도']
    
    title = dms.loc[i, 'DMS 지역명']
    
    color = 'blue'
    
    # 지도에 동그라미로 데이터 찍기
    folium.CircleMarker([sub_lat, sub_long], color=color, radius=7, tooltip=title).add_to(m)

# 한글이 나오지 않는 오류로 html로 trouble shooting
m.save('dms.html')

m
```
