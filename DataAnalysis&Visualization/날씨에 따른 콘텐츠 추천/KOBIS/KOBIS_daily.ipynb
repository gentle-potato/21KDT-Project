```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('KOBIS_daily_191231~210625.csv')

data
```

```python
type(data)
```

```python
data["daily_audience"] = 0

data
```

```python
for i in range(len(data)):
    data['title'][i] = data['title'][i][:-4]

data
```

```python
data = np.array(data)

data
```

```python
movie_dict = {}

for i in range(len(data)):
    list = data[i]
    if list[2] not in movie_dict:
        movie_dict[f'{list[2]}'] = list[10]
        list[11] = list[10]
    else:
        list[11] = list[10] - movie_dict[f'{list[2]}']
        movie_dict[f'{list[2]}'] = list[10]

data
```

```python
data = pd.DataFrame(data,
                   columns = ['daily_rank', 'poster', 'title', 'week', 'release_date', 'country', 'genre', 'director', 'act', 'runtime', 'audience', 'daily_audience'])

data
```
