## 공휴일 데이터 합치기

```python
df_hol = pd.read_csv('국가공휴일.csv', encoding='cp949')
df_hol
```

```python
df_hol.info()
```

```python
def todatetime(x):
    year = str(x[1])
    month = x[2]
    day = x[3]
    if month < 10:
        month = '0' + str(month)
    else:
        month = str(month)
    
    if day < 10:
        day = '0' + str(day)
    else:
        day = str(day)
    
    return year+'-'+month+'-'+day
```

```python
df_hol['base_date'] = df_hol.apply(todatetime, axis=1)
```

```python
df_hol['holiday'] = True
```

```python
df_hol
```

```python
df2 = pd.merge(df, df_hol[['base_date', 'holiday']], on='base_date', how='left')
```

```python
df2.fillna(False, inplace=True)
df2
```

## 평일 데이터 생성 후 합치기

```python
start_date = pd.to_datetime('2018-01-01') # 시작 날짜
end_date = pd.to_datetime('2021-06-30') # 마지막 날짜
 
dates = pd.date_range(start_date,end_date,freq='B') # 주말이 아닌 날짜만 생성
dates
```

```python
weekdays = pd.DataFrame(data={'weekend':[False for _ in range(len(dates))]}, index=dates)
```

```python
weekdays.index.name = 'base_date'
```

```python
weekdays.reset_index(inplace=True)
weekdays
```

```python
weekdays['base_date'] = weekdays['base_date'].apply(lambda x: x.strftime('%Y-%m-%d'))
```

```python
weekdays
```

```python
df2 = pd.merge(df2, weekdays, on='base_date', how='left')
df2
```

```python
df2.fillna(True, inplace=True)
df2
```

```python
df2['dayoff'] = (df2['holiday'] | df2['weekend'])
```

```python
df2.replace({True:1, False:0}, inplace=True)
```

## 카드 소비 데이터 합치기
- 관심 있는 카테고리인 배달, 식품, 농축수산물, 마트/슈퍼마켓의 데이터만 합치기

```python
df_card = pd.read_csv('01_제공데이터_update(210811)/04_음식관련 카드소비_CARD_SPENDING.CSV', encoding='cp949')
```

```python
df_card = df_card[df_card['mct_cat_nm'].isin(['배달', '식품', '농축수산물', '마트/슈퍼마켓', '간식'])]
```

```python
df_card
```

### 카테고리 column 생성

```python
for cate in ['배달', '식품', '농축수산물', '마트/슈퍼마켓', '간식']:
    df_card['use_cnt'+cate] = df_card[['mct_cat_nm', 'use_cnt']].apply(lambda x: x[1] if x[0]==cate else 0, axis=1)
    df_card['use_amt'+cate] = df_card[['mct_cat_nm', 'use_amt']].apply(lambda x: x[1] if x[0]==cate else 0, axis=1)
```

```python
df_card
```

```python
df_card2 = df_card.groupby(['base_date', 'city', 'emd_nm']).sum().reset_index()
```

```python
df_card2.drop(columns=['use_cnt', 'use_amt'], inplace=True)
```

```python
df_card2
```

```python
df = pd.read_csv("전체 데이터 통합2.csv", encoding='cp949')
```

```python
df3 = pd.merge(df, df_card2[['base_date', 'city', 'emd_nm', 'use_amt간식']], on=['base_date', 'city', 'emd_nm'], how='left')
```

```python
df3
```

```python
df3.to_csv('전체데이터통합3.csv', encoding='cp949', index=None)
```

### 각 읍면동의 구역

```python
df3['emd_nm'].unique()
```

```python
len(df3['emd_nm'].unique())
```

```python
west = ['한림읍', '한경면', '대정읍']
north = ['애월읍', '외도동', '건입동', '노형동', '도두동', '봉개동', '삼도1동', '삼도2동',
        '삼양동', '아라동', '연동', '오라동', '용담1동', '용담2동', '이도1동',
        '이도2동', '이호동', '일도1동', '일도2동', '화북동', '조천읍']
east = ['성산읍', '표선면', '구좌읍']
south = ['남원읍', '대륜동', '대천동', '동홍동', '서홍동', '송산동', '안덕면', '영천동',
        '예래동', '정방동', '중문동', '중앙동', '천지동', '효돈동']
```

```python
len(west+north+east+south)
```

```python
def allo(x):
    if x in west:
        return '고산'
    elif x in north:
        return '제주'
    elif x in east:
        return '성산'
    else:
        return '서귀포'
```

```python
df3['지점명'] = df3['emd_nm'].apply(allo)
```
