```python
import pandas as pd
```

```python
df = pd.read_csv('01_제공데이터_update(210811)/03_거주인구_RESIDENT_POP.CSV', encoding='cp949')
```

```python
df
```

```python
def todatetime(x):
    year = str(x[0])
    month = x[1]
    if month < 10:
        month = '0' + str(month)
    else:
        month = str(month)
    
    return year+'-'+month
```

```python
df.insert(0, 'base_date', df.apply(todatetime, axis=1))
```

```python
df = df[['base_date', 'city', 'emd_nm', 'sex', 'resid_reg_pop', 'foreign_pop']]
df
```

```python
for sex in ['남성', '여성']:
    df['내국인거주'+sex] = df[['sex', 'resid_reg_pop']].apply(lambda x: x[1] if x[0]==sex else 0, axis=1)
    df['외국인거주'+sex] = df[['sex', 'foreign_pop']].apply(lambda x: x[1] if x[0]==sex else 0, axis=1)
```

```python
df
```

```python
df2 = df.groupby(['base_date', 'city', 'emd_nm']).sum().iloc[:, 2:].reset_index().rename(columns={'base_date':'month'})
df2
```

```python
df_base = pd.read_csv('전체데이터통합.csv', encoding='cp949')
```

```python
df_base
```

```python
df_base['month'] = df_base['base_date'].str[:7]
df_base
```

```python
df_all = pd.merge(df_base, df2, on=['month', 'city', 'emd_nm'], how='left')
```
