# **Exploratory Data Analysis**

```python
import pandas as pd
import numpy as np

import folium

import warnings
warnings.filterwarnings('ignore')
# pd.set_option('display.max_rows', None)
```

## **1. 신규 아파트 단지 정보**

```python
new = pd.read_csv('../../../../Source/2021 빅콘테스트/2021_빅콘테스트_지역활성화/금융-부동산/부동산 데이터_지인플러스/신규 아파트 단지 정보.csv', encoding='utf-8')

new.head()
```

```python
for i in range(len(new['LGDNG_CD'])):
    new['LGDNG_CD'][i] = str(new['LGDNG_CD'][i])[0:4]

new.tail()
```

```python
new['LGDNG_CD'].value_counts()
```

군산 16개 / 목포 18개 / 통영 11개

```python
new_apt = new.drop(['BLDID_ID', 'BUN', 'JI', 'CMPL_PARG_DATE'], axis=1)

new_apt.head()
```

```python
type(new_apt.TOTHH_CT[0])
```

```python
new_apt.TOTHH_CT[0]*0.01
```

```python
GunSan_new = new[new['LGDNG_CD'] == 4513]
GunSan_new = GunSan_new.reset_index()
GunSan_new.drop('index', axis=1, inplace=True)
MokPo_new = new[new['LGDNG_CD'] == 4611]
MokPo_new = MokPo_new.reset_index()
MokPo_new.drop('index', axis=1, inplace=True)
TongYeong_new = new[new['LGDNG_CD'] == 4822]
TongYeong_new = TongYeong_new.reset_index()
TongYeong_new.drop('index', axis=1, inplace=True)
```

### **① GunSan**

```python
# 지도 새로 띄우기
m_GunSan_new = folium.Map([np.mean(GunSan_new['LA']), np.mean(GunSan_new['LNGT'])], zoom_start=12)

for i in GunSan_new.index:
    sub_lat = GunSan_new.loc[i, 'LA']
    sub_long = GunSan_new.loc[i, 'LNGT']
    
    title = GunSan_new.loc[i, 'HSCPX_NM']
    
    color = 'red'
    
    # 지도에 동그라미로 데이터 찍기
    folium.CircleMarker([sub_lat, sub_long], color=color, radius=GunSan_new.TOTHH_CT[i]*0.02, tootip=title, fill=True).add_to(m_GunSan_new)

m_GunSan_new.save('apt_GunSan_new.html')

m_GunSan_new
```

### **② MokPo**

```python
# 지도 새로 띄우기
m_MokPo_new = folium.Map([np.mean(MokPo_new['LA']), np.mean(MokPo_new['LNGT'])], zoom_start=12)

for i in MokPo_new.index:
    sub_lat = MokPo_new.loc[i, 'LA']
    sub_long = MokPo_new.loc[i, 'LNGT']
    
    title = MokPo_new.loc[i, 'HSCPX_NM']
    
    color = 'green'
    
    # 지도에 동그라미로 데이터 찍기
    folium.CircleMarker([sub_lat, sub_long], color=color, radius=MokPo_new.TOTHH_CT[i]*0.02, tootip=title, fill=True).add_to(m_MokPo_new)

m_MokPo_new.save('apt_MokPo_new.html')

m_MokPo_new
```

### **③ TongYeong**

```python
# 지도 새로 띄우기
m_TongYeong_new = folium.Map([np.mean(TongYeong_new['LA']), np.mean(TongYeong_new['LNGT'])], zoom_start=12)

for i in TongYeong_new.index:
    sub_lat = TongYeong_new.loc[i, 'LA']
    sub_long = TongYeong_new.loc[i, 'LNGT']
    
    title = TongYeong_new.loc[i, 'HSCPX_NM']
    
    color = 'blue'
    
    # 지도에 동그라미로 데이터 찍기
    folium.CircleMarker([sub_lat, sub_long], color=color, radius=TongYeong_new.TOTHH_CT[i]*0.02, tootip=title, fill=True).add_to(m_TongYeong_new)

m_TongYeong_new.save('apt_TongYeong_new.html')

m_TongYeong_new
```
